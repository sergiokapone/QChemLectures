```python
from pyscf import gto, scf
import numpy as np
```

# Рахуємо атом водню

```python
H = gto.M(atom = 
           '''H 0 0 0 
           ''',
           charge = 0,
           spin = 1, # spin = N_alpha - N_beta 
               basis = 'ccpvdz')
```

# Показуємо число $\alpha$- і $\beta$-електронів ($N_{\alpha}$, $N_{\beta}$)

```python
H.nelec
```

# Запускаємо ітераційну процедуру (SCF)

```python
hm = scf.UHF(H).run()
```

# Обчислюємо повну енергію атома H

За теорією Шредінгера (або Бора) енергія основного стану атома H $E = - \frac{Z^2}{2} = -\frac12 = -0.5$~h

```python
hm.e_tot
```

```python
hm.energy_nuc()
```

# Рахуємо атом гелію

Експериментальне значення $E = -2.9037$ h

```python
He = gto.M(atom = 
           '''He 0 0 0 
           ''',
           charge = 0,
           spin = 2, # spin = 1 + 1
               basis = 'ccpvdz')
```

```python
hm = scf.UHF(He).run()
```

```python
hm.e_tot
```
