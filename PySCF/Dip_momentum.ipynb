```python
from pyscf import gto, scf
import numpy as np
```

# Задаємо геометрію і базис

```python
basis = '6-31g*'
```

```python
mol = gto.M(atom= '''O   0.000000   0.000000   0.117176
                     H   0.000000   0.755453  -0.468706
                     H   0.000000  -0.755453  -0.468706
                     ''',
            charge=0,
            spin=0,
            basis=basis,
            unit='Angstrom')
```

# Запускаємо розрахунок методом Restricted Hartree Fock

```python
mf = scf.RHF(mol)   # mean-field объект для HF
```

# Отримуємо значення енергії

```python
mf.kernel()
```

```python
dipole = mf.dip_moment()
dipole_magnitude = np.linalg.norm(dipole)
dipole_magnitude
```

# Оптимізуємо геометрію

```python
from pyscf import geomopt
mol_geom_opt = geomopt.optimize(mf)
```

# Створюємо нову молекулу на основі оптимізованої геометрії

```python
from pyscf import lib
bohr_to_ang = lib.param.BOHR
coords_bohr = mol_geom_opt.atom_coords()
coords_ang  = coords_bohr * bohr_to_ang
symbols = [mol_geom_opt.atom_symbol(i) for i in range(mol_geom_opt.natm)]
symbols
with open("optimized.xyz", "w") as f:
    f.write(f"{mol_geom_opt.natm}\n")
    f.write("Optimized geometry from PySCF\n")
    for s, (x, y, z) in zip(symbols, coords_ang):
        f.write(f"{s:2s} {x:15.8f} {y:15.8f} {z:15.8f}\n")
```

```python
mol_opt = gto.M(atom='optimized.xyz',
            charge=0,
            spin=0,
            basis=basis,
            unit='Angstrom')
```

```python
mf_opt = scf.RHF(mol_opt)   # mean-field объект для HF
mf_opt.kernel()
```

# Обчислюємо дипольний момент

```python
dipole = mf_opt.dip_moment()
```

```python
dipole_magnitude = np.linalg.norm(dipole)
dipole_magnitude
```
