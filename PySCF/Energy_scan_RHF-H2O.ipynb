# Молекула H2O



# Розрахунки

```python
from pyscf import gto, scf, cc
import numpy as np
import matplotlib.pyplot as plt
```

```python
basis = 'cc-PVDZ'
```

## Сканування ППЕ методом HF

```python
angle = []
E = []
```

```python
d = 0.9584
for t in np.linspace(45, 315, 100):
    mol = gto.M(atom=f'''O 0              0            0
                         H {d}            0            0
                         H {d*np.cos(t*np.pi/180)} {d*np.sin(t*np.pi/180)} 0
                      ''',
                basis=basis,
                unit = 'Angstrom'
               )
    mf = scf.RHF(mol)
    mf.verbose=0
    e = mf.kernel()
    angle.append(t)
    E.append(e)
```

```python
# малюємо графік
plt.plot(angle, E, marker="o",)
plt.xlabel("Кут H-O-H (градуси)")
plt.ylabel("Енергія (a.u.)")
plt.title(r'Потенціальна крива для молекули $\mathrm{{H}}_2\mathrm{{O}}$')
plt.grid(True)
plt.show()
```
