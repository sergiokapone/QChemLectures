```python
from pyscf import gto, scf
```

# Задаємо геометрію і базис

```python
mol = gto.M(atom='''
                 O       -2.60858        0.82218        0.00000
                 H       -1.63858        0.82218        0.00000
                 H       -2.93191        1.69020       -0.28792
                 ''',
            charge=0,
            spin=0,
            basis='sto3g',
            unit='Angstrom')
```

# Запускаємо розрахунок методом Restricted Hartree Fock

```python
mf = scf.RHF(mol)   # mean-field объект для HF
```

```python
mf.kernel()
```
