```python
from pyscf import gto, scf
```

# Задаємо геометрію і базис

```python
mol = gto.M(atom='''
                 He       0        0        0
                 ''',
            charge=0,
            spin=0,
            basis='sto3g',
            unit='Angstrom')
```

# Запускаємо розрахунок методом Restricted Hartree Fock

```python
mf = scf.RHF(mol)  # mean-field объект для HF
```

```python
E = mf.kernel()
```
