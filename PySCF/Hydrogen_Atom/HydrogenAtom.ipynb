# Атома водню в квантовій механіці

Атом водню — це найпростіша система в квантовій механіці, яка складається з одного протона (ядра) і одного електрона. Його дослідження дозволяє зрозуміти основи квантової механіки, спектроскопії та електронних орбіталей.

---

## 1. Основне рівняння

Повна хвильова функція електрона $\psi(\mathbf{r})$ описується **рівнянням Шредінгера**:

### 1.1. Стаціонарне рівняння Шредінгера

$$
\hat{H} \psi(\mathbf{r}) =E_n \psi(\mathbf{r})
$$

де $\hat{H}$ — гамільтоніан системи.

### 1.2. Гамільтоніан атома водню (в системі центра мас)

Для електрона у кулонівському полі протона:

$$
\hat{H} = -\frac{\hbar^2}{2m_e} \nabla^2 - \frac{e^2}{r}
$$

- $m_e$ — маса електрона  
- $e$ — заряд електрона  
- $r = |\mathbf{r}|$ — відстань електрона від ядра  

---

## 2. Розв'язок рівняння Шредінгера

Оскільки потенціал центральний, зручно використовувати **сферичні координати** $(r, \theta, \phi)$. Хвильова функція розкладається у вигляді:

$$
\Psi_{n l m}(r, \theta, \phi) = R_{n l}(r) \, Y_{l}^{m}(\theta, \phi)
$$

- $R_{n l}(r)$ — **радіальна частина**  
- $Y_l^m(\theta, \phi)$ — **сферична гармоніка** (кутова частина)  
- $n = 1, 2, 3, \dots$ — головне квантове число  
- $l = 0, 1, \dots, n-1$ — орбітальне (кутове) квантове число  
- $m = -l, \dots, l$ — магнітне квантове число  

---

## 3. Енергетичні рівні для воднеподібних атомів

Енергія електрона залежить тільки від головного квантового числа $n$ та заряду ядра $Z$:



### У системі СГС (с, см, г):

$$
E_n = - \frac{m_e e^4 Z^2}{2 \hbar^2 n^2} \quad \text{(в ергах, де $e$ у статкулонах, $m_e$ в грамах, $\hbar$ в ерг·с)}
$$

- $m_e$ — маса електрона ($9.109 \cdot 10^{-28} \, \text{г}$)  
- $e$ — заряд електрона ($4.8 \cdot 10^{-10} \, \text{Фр}$)  
- $\hbar$ — зведена стала Планка ($1.055 \cdot 10^{-27} \, \text{ерг·с}$)  
- $Z$ — заряд ядра (1 для водню, 2 для He$^+$, 3 для Li$^{2+}$ і т.д.)  

### У атомних одиницях:

$$
E_n = - \frac{Z^2}{2 n^2}
$$

- $n=1$ — основний стан  
- $n>1$ — збуджені стани  
- Кожен рівень $n$ має **виродженість** $n^2$  
- Для воднеподібних атомів енергія зростає з $Z^2$ (сильніше зв'язування електрона)
---

## 4. Радіальна частина $R_{nl}(r)$

Ра́діальна хвильова функція задається **поліномом Лагерра**:

$$
R_{nl}(r) = \sqrt{\left( \frac{2}{n a_0} \right)^3 \frac{(n-l-1)!}{2n [(n+l)!]}} \, e^{-r/(n a_0)} \left(\frac{2r}{n a_0}\right)^l L_{n-l-1}^{2l+1} \left( \frac{2r}{n a_0} \right)
$$

- $a_0 = \frac{\hbar^2}{m_e e^2}$ — **борадівський радіус** (~0.529 Å)  
- $L_{n-l-1}^{2l+1}$ — **асоційовані поліноми Лагерра**

---

## 5. Кутова частина — сферичні гармоніки

Сферичні гармоніки $Y_l^m(\theta, \phi)$ описують **орієнтацію орбіталі в просторі**:

$$
Y_l^m(\theta, \phi) = (-1)^m \sqrt{\frac{(2l+1)}{4\pi} \frac{(l-m)!}{(l+m)!}} \, P_l^m(\cos\theta) \, e^{i m \phi}
$$

- $P_l^m$ — **асоційовані поліноми Лежандра**  
- Для p-орбіталей ($l=1$):  
  $$
  Y_1^{-1} = \sqrt{\frac{3}{8\pi}} \sin\theta \, e^{-i\phi}, \quad
  Y_1^{0} = \sqrt{\frac{3}{4\pi}} \cos\theta, \quad
  Y_1^{1} = -\sqrt{\frac{3}{8\pi}} \sin\theta \, e^{i\phi}
  $$

- Реальні p-орбіталі ($p_x, p_y, p_z$):

$$
p_x = \frac{1}{\sqrt{2}} (Y_1^{-1} - Y_1^1) = \sqrt{\frac{3}{4\pi}} \sin\theta \cos\phi
$$

$$
p_y = \frac{i}{\sqrt{2}} (Y_1^{-1} + Y_1^1) = \sqrt{\frac{3}{4\pi}} \sin\theta \sin\phi
$$

$$
p_z = Y_1^0 = \sqrt{\frac{3}{4\pi}} \cos\theta
$$

---

## 6. Спін електрона

Електрон має **спін** $s = 1/2$ із проекціями $m_s = \pm 1/2$. Повна хвильова функція включає і спінову частину:

$$
\Psi_{\text{повна}} = \Psi_{n l m}(\mathbf{r}) \, \chi_{m_s}
$$

---

## 7. Висновки

1. Енергетичні рівні атома водню визначаються тільки головним квантовим числом $n$.  
2. Радіальна частина визначає "відстань" електрона від ядра, кутова — форму орбіталі.  
3. p-орбіталі мають форму гантелі, d- і f-орбіталі — складніші форми.  
4. Теорія атома водню є фундаментом для багатьох квантових систем, спектроскопії та хімії.

```python
# Створимо список n і l для прикладу
levels = [
    (2, 0),
    (2, 1),
    (3, 0),
    (3, 1),
    (3, 2)
]

# Створюємо Markdown/LaTeX-таблицю
table_header = r"| $(n,l)$ | $R_{nl}(r,Z)$ |" + "\n|---|---|"
table_rows = []

for n, l in levels:
    R = R_nl(n, l, r, Z)  # символьна функція
    # Перетворюємо на LaTeX для рендерингу
    R_latex = latex(R)
    row = f"| ({n},{l}) | $${R_latex}$$ |"
    table_rows.append(row)

# Об'єднуємо все
table_md = table_header + "\n" + "\n".join(table_rows)

# Виводимо як Markdown
from IPython.display import Markdown
display(Markdown(table_md))
```

```python
R10 = lambdify(r, R_nl(1, 0, r, 1), 'numpy')
R20 = lambdify(r, R_nl(2, 0, r, 1), 'numpy')
R21 = lambdify(r, R_nl(2, 1, r, 1), 'numpy')
R30 = lambdify(r, R_nl(3, 0, r, 1), 'numpy')
R31 = lambdify(r, R_nl(3, 1, r, 1), 'numpy')
R32 = lambdify(r, R_nl(3, 2, r, 1), 'numpy')
```

```python
r = np.linspace(0, 15, 100)
```

```python
# Создаём сетку 2x3 подграфиков
fig, axes = plt.subplots(2, 3, figsize=(15, 10))
axes = axes.flatten() # превращаем в одномерный массив для удобства


# Список функций и их названий
orbitals = [(R10, '$R_{10}$'),
(R20, '$R_{20}$'),
(R21, '$R_{21}$'),
(R30, '$R_{30}$'),
(R31, '$R_{31}$'),
(R32, '$R_{32}$')]


# Рисуем каждый график на отдельной оси
for ax, (R, label) in zip(axes, orbitals):
    ax.plot(r, R(r), label=label)
    # ax.set_ylim(0, 0.05)
    ax.set_yticks([])
    ax.spines['left'].set_position('zero')
    ax.spines['bottom'].set_position('zero')
    ax.spines['top'].set_color('none')
    ax.spines['right'].set_color('none')
    ax.xaxis.set_ticks_position('bottom')
    ax.yaxis.set_ticks_position('left')
    ax.legend()


plt.tight_layout() # подгонка расстояний между подграфиками
plt.show()
```
