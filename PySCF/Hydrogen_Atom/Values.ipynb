```python
from sympy import symbols, pi, exp, sqrt, I, oo, integrate, diff, conjugate, sin, cos, simplify
from sympy.physics.hydrogen import Psi_nlm
from sympy.abc import r, phi,theta, Z, n, l, m
import numpy as mp
import matplotlib.pyplot as plt
```

```python
Psi = Psi_nlm(3,1,0, r, phi, theta, 3)
```

```python
Psi_c = conjugate(Psi)
```

```python
result = integrate(
    Psi_c * r * Psi * r**2 * sin(theta),  
    (r, 0, oo),
    (theta, 0, pi),  # θ от 0 до π 
    (phi, 0, 2*pi)   # φ от 0 до 2π
)
result
```

```python

r, theta, phi = symbols('r theta phi', real=True, positive=True)
hbar = symbols('hbar', positive=True, real=True)

n, l, m = 1, 0, 0
psi_100 = Psi_nlm(n, l, m, r, theta, phi)
p_r_psi = -I * hbar * diff(psi_100, r)
p_theta_psi = -I * hbar * (1/r) * diff(psi_100, theta)
p_phi_psi = -I * hbar * (1/(r*sin(theta))) * diff(psi_100, phi)




integrand_r = conjugate(psi_100) * p_r_psi * r**2 * sin(theta)


integrate(integrand_r, (r, 0, oo), (theta, 0, pi), (phi, 0, 2*pi))
```
