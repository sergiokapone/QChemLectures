```python
import sympy as sp
from sympy import symbols, pi, exp, sqrt, I, oo, integrate, diff, conjugate, simplify, sin, cos
from sympy.physics.hydrogen import Psi_nlm, R_nl
import matplotlib.pyplot as plt
import numpy as np
```

# Обчислення середнього значення імпульсу для атома водню

## Знайдемо $\langle \hat{p} \rangle$ для основного стану атома водню

```python
Z, r, theta, phi = symbols('Z r theta phi', real=True, positive=True)
```

```python
# Хвильова функція основного стану (1s)
n, l, m = 1, 0, 0
psi_100 = Psi_nlm(n, l, m, r, theta, phi, Z)
psi_100
```

## Оператор імпульсу в сферичних координатах

```python
# Дія оператора радіального імпульсу
T_psi = 1/2 * diff(r ** 2 * psi_100, r)
```

```python
T = integrate(4 * pi *  psi_100 * T_psi , (r, 0, oo))
T
```

```python
U = integrate(psi_100 * (-1 / r) * psi_100 * 4 * pi* r ** 2, (r, 0, oo))
U
```

```python
E = T + U
E
```
