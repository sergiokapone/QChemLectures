```python
# ---------------------

from sympy import *
from sympy.physics.hydrogen import Psi_nlm, R_nl, E_nl
from sympy.abc import phi,theta,  n, l, m

# ---------------------
import numpy as mp
import matplotlib.pyplot as plt
```

# Обчислити інтеграл $\left\langle \psi_{100} | \psi_{100} \right\rangle = \iiint\limits_V \psi^*_{100} \psi_{100} dV$

```python
r = symbols('r', positive=True)
Z = symbols('Z', positive=True)
```

```python
psi_100 = Psi_nlm(1,0, 0, r, theta, phi, Z)
psi_100 * psi_100
```

```python
conjugate(psi_100)
```

```python
integrate(conjugate(psi_100) * psi_100 * 4 * pi * r ** 2 , (r, 0, oo))
```

# Середня потенціальна енергія
$\left\langle \psi_{100} | \hat{U} |\psi_{100} \right\rangle = \iiint\limits_V \psi^*_{100} \left(-\frac{Z}{r} \right)  \psi_{100} dV $

```python
U = (-Z/r)
```

```python
U = integrate(conjugate(psi_100) * U * psi_100 * 4 * pi * r ** 2, (r, 0, oo))
U
```

# Середня кінетична енергія $\left\langle T \right\rangle = \left\langle \psi_{100} | \hat{T} |\psi_{100} \right\rangle = \iiint\limits_V \psi^*_{100} \left( - \frac12 \nabla^2  \psi_{100} \right)  dV $

```python
T_psi_100 = - 1/2 * 1/r ** 2 * diff(r ** 2 * diff(psi_100, r), r)
```

```python
T = integrate(conjugate(psi_100) * T_psi_100 * 4 * pi * r ** 2, (r, 0, oo))
T
```

# Середня повна енергія $\left\langle E \right\rangle = \left\langle T \right\rangle + \left\langle U \right\rangle$

```python
E = T + U # Середня
E
```

```python
E_nl(1, Z) # точно по теорії 
```

# Середній радіус атома $\left\langle r \right\rangle  = \left\langle \psi_{100} | \hat{r} |\psi_{100} \right\rangle = \iiint\limits_V \psi^*_{100} r  \psi_{100}  dV   = 4\pi \int\limits_0^{\infty} \psi^*_{100} r \psi_{100}  \cdot r^2 dr$

```python
4 * pi * integrate(conjugate(psi_100) * r * psi_100 *  r ** 2, (r, 0, oo))
```
