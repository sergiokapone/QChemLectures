```python
from pyscf import gto, scf, ci
import numpy as np
```

```python
mol = gto.M(
          atom = f'H 0 0 0; H 0 0 0.741',
          basis = 'cc-PVDZ'
         )
```

```python
def make_cisd(mf):
    obj = ci.CISD(mf)
    obj.nstates = 15
    return obj
```

```python
ci_scaner = gto.M(verbose=0).apply(scf.RHF).apply(make_cisd).as_scanner()
```

```python
ci_scaner(mol)
```
