# Молекулярні орбіталі та радіальна функція розподілу (RDF)

## 1. Молекулярні орбіталі (MO)

У методах типу Hartree-Fock (HF) або DFT хвильова функція електрона виражається через **молекулярні орбіталі**:

$$
\psi_j(\mathbf r) = \sum_{\mu} C_{\mu j} \, \chi_\mu(\mathbf r)
$$

де:

- $\psi_j(\mathbf r)$ — $j$-та молекулярна орбіталь,  
- $C_{\mu j}$ — коефіцієнти розкладу МО по атомних орбіталях (AO),  
- $\chi_\mu(\mathbf r)$ — атомні орбіталі (лінійна комбінація гауссових функцій).  


## 2. Електронна густина

Електронна густина системи визначається через молекулярні орбіталі (МО):

$$
\rho(\mathbf r) = \sum_{j} n_j \, |\psi_j(\mathbf r)|^2,
$$

де $n_j$ — число електронів на $j$-й МО.

Підставляючи вираз для МО через атомні орбіталі (АО):

$$
\psi_j(\mathbf r) = \sum_\mu C_{\mu j} \chi_\mu(\mathbf r),
$$

отримуємо:

$$
\rho(\mathbf r) = \sum_j n_j 
\left( \sum_\mu C_{\mu j} \chi_\mu(\mathbf r) \right)
\left( \sum_\nu C_{\nu j} \chi_\nu(\mathbf r) \right).
$$

Розкриваючи добутки, маємо:

$$
\rho(\mathbf r) = \sum_{\mu,\nu} 
\left( \sum_j n_j C_{\mu j} C_{\nu j} \right)
\chi_\mu(\mathbf r)\chi_\nu(\mathbf r).
$$

Визначимо **матрицю густини у базисі АО**:

$$
D_{\mu\nu} = \sum_j n_j C_{\mu j} C_{\nu j},
$$

тоді:

$$
\rho(\mathbf r) = \sum_{\mu,\nu} D_{\mu\nu} \,
\chi_\mu(\mathbf r)\chi_\nu(\mathbf r).
$$

---

### Матрична форма

Позначимо вектор АО-значень:

$$
\boldsymbol{\chi}(\mathbf r) = 
\begin{bmatrix}\chi_1(\mathbf r) \\ \chi_2(\mathbf r) \\ \vdots \\ \chi_{N_{\text{AO}}}(\mathbf r)\end{bmatrix}.
$$

Тоді густина компактно записується як:

$$
\rho(\mathbf r) \;=\; \boldsymbol{\chi}(\mathbf r)^\top \, \mathbf D \, \boldsymbol{\chi}(\mathbf r),
$$

де матриця густини виражається через коефіцієнти МО $C$ та діагональну матрицю заповненостей $N$:

$$
\mathbf D \;=\; C \, N \, C^\top
\quad\text{(або $C \, N \, C^\dagger$ для комплексних МО).}
$$

---

## 3. Радіальна функція розподілу (RDF)

Для атома або сферично симетричної системи часто цікавить **радіальна густина** $\rho(r)$, усереднена по кутам:

$$
\rho(r) = \frac{1}{4\pi} \int |\psi(\mathbf r)|^2 \, d\Omega
$$

де $d\Omega$ — елемент поверхні сфери.  

**Радіальна функція розподілу (RDF):**

$$
P(r) = 4 \pi r^2 \, \rho(r)
$$

- Показує, як електронна густина розподілена по радіусах,  
- Піки $P(r)$ відповідають електронним оболонкам (наприклад, 1s, 2s у літію).  

**Обчислення на сітці:**

1. Створюється радіальна сітка $r_i$ навколо ядра.  
2. Обчислюється $\psi_j(\mathbf r_i)$ через розклад по AO.  
3. Густина: $\rho(\mathbf r_i) = \sum_j n_j |\psi_j(\mathbf r_i)|^2$.  
4. Усереднення по кутам (для s-симетрії можна брати пряму лінію):  
   $\rho(r_i) = \langle \rho(\mathbf r_i) \rangle_\Omega$.  
5. RDF: $P(r_i) = 4 \pi r_i^2 \rho(r_i)$.  


```python
from pyscf import gto, scf
from pyscf.tools import molden
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
basis  = 'sto3g'
```

```python
# Задаём молекулу
mol = gto.M(atom='''
            Ar  0.0000   0.000000   0.000000
            ''',
    basis = basis,
    charge=0,
    spin=0,
    verbose=0
)
```

```python
# RHF расчёт
mf = scf.RHF(mol)
mf.conv_tol = 1e-10
mf.max_cycle = 100
mf.kernel()
```

```python
# Извлекаем данные
mo_energy = mf.mo_energy
mo_occ = mf.mo_occ
mo_coeff = mf.mo_coeff
ao_labels = mol.ao_labels()
```

```python
mo_occ
```

```python
columns_multiindex = pd.MultiIndex.from_arrays([
    range(len(ao_labels)), 
    mo_energy,
    mo_occ,
], names=['MO_Index', 'Orbiral Energy', 'OCC'])
```

```python
df = pd.DataFrame(mo_coeff, 
                  columns=columns_multiindex,
                  index=ao_labels)
```

```python
def line_directions(n_points, r_max, direction):
    """
    Генерує n_points точок вздовж прямої у напрямку `direction`, 
    від 0 до r_max.
    
    direction: масив [dx, dy, dz], буде нормалізовано.
    """
    direction = np.array(direction)
    direction = direction / np.linalg.norm(direction)  # нормалізація
    r = np.linspace(0, r_max, n_points)              # відстані вздовж прямої
    coords = np.outer(r, direction)                  # (n_points, 3)
    return coords, r


n_points = 1000
r_max = 5.0
direction = [0, 1, 0]  # напрямок в просторі

coords_line, r_line = line_directions(n_points, r_max, direction)
```

```python
# Создаём одномерную сетку вдоль оси x
ao_values = mol.eval_gto('GTOval_sph', coords_line)
```

# Обчислюємо МО

$$
\psi_j(\mathbf r) = \sum_\mu C_{\mu j} \chi_\mu(\mathbf r),
$$

або в матричному вигляді

$$
\boldsymbol{\psi}(\mathbf r) = \boldsymbol{\chi}(\mathbf r)^\top \, \mathbf{C},
$$

$$
\boldsymbol{\chi}(\mathbf r) =
\begin{bmatrix}
\chi_1(\mathbf r) \\[1mm]
\chi_2(\mathbf r) \\[1mm]
\vdots \\[1mm]
\chi_{N_{\rm AO}}(\mathbf r)
\end{bmatrix}, 
\quad
\mathbf{C} =
\begin{bmatrix}
C_{11} & C_{12} & \cdots & C_{1 N_{\rm MO}} \\
C_{21} & C_{22} & \cdots & C_{2 N_{\rm MO}} \\
\vdots & \vdots & \ddots & \vdots \\
C_{N_{\rm AO} 1} & C_{N_{\rm AO} 2} & \cdots & C_{N_{\rm AO} N_{\rm MO}}
\end{bmatrix}.
$$

```python
mo_values = ao_values @ mo_coeff 
```

```python
plt.figure(figsize=(10, 6))
for i in range(len(np.where(mo_occ > 0)[0])):
    plt.plot(r_line, mo_values[:, i], label=f'MO {i}')
plt.xlabel('r (Borh)')
plt.ylabel('Значение МО')
plt.title('Молекулярные орбитали Li')
plt.legend()
plt.grid()
plt.show()
```

```python
# електронна гутина
density = (np.abs(mo_values)**2) @ mo_occ 

# N = np.diag(mo_occ)           # матриця N
# D = mo_coeff @ N @ mo_coeff.T # матриця густини в AO-базисі

# #density = np.sum((ao_values @ mo_coeff) * (ao_values @ (mo_coeff @ N)), axis=1)
# density = np.einsum("im, mn, in -> i", ao_values, D, ao_values)
```

```python
plt.plot(r_line, 4 * np.pi * r_line ** 2 * density)
plt.xlabel('r (Borh)')
plt.ylabel(r'$4\pi r^2 \rho(r)$')
plt.title('RDF')
plt.show()
```
