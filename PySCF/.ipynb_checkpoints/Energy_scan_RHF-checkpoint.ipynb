```python
from pyscf import gto, scf, cc
import numpy as np
import matplotlib.pyplot as plt
```

```python
basis = '6-31G*'
```

```python
# мелкий шаг около минимума
R1 = np.arange(0.25, 1.5, 0.05)  # плотная сетка

# большой шаг вдали от минимума
R2 = np.arange(1.5, 5.0, 0.25)   # редкая сетка

# объединяем в один массив
distances = np.concatenate([R1, R2])
```

# Сканування ППЕ звичайним методом RHF

```python
R = []
E = []
```

```python
scanner = gto.M(verbose=0).apply(scf.RHF).as_scanner()
for r in distances:
    e = scanner(gto.M(atom=f"H 0 0 0; H 0 0 {r}", basis=basis))
    R.append(r)
    E.append(e)
```

```python
R
```

```python
# рисуем график
plt.plot(R, E, marker="o")
plt.xlabel("Расстояние H-H (Å)")
plt.ylabel("Энергия (a.u.)")
plt.title("Потенциальная кривая H2")
plt.grid(True)
plt.show()
```

# Сканування енергії методом Coupled Clusters

```python
R = []
E = []
```

```python
cc_scanner = gto.M(verbose=0).apply(scf.RHF).apply(cc.CCSD).as_scanner()
for r in distances:
    e = cc_scanner(gto.M(atom=f"H 0 0 0; H 0 0 {r}", basis=basis))
    R.append(r)
    E.append(e)
```

```python
# рисуем график
plt.plot(R, E, marker="o")
plt.xlabel("Расстояние H-H (Å)")
plt.ylabel("Энергия (a.u.)")
plt.title("Потенциальная кривая H2")
plt.grid(True)
plt.show()
```
