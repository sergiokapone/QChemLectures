```python
import numpy as np
import matplotlib.pyplot as plt
from sympy.physics.hydrogen import R_nl,  Psi_nlm
from sympy.abc import r, phi,theta, Z
from sympy import lambdify
```

```python
Psi100 = Psi_nlm(1, 0, 0, r, phi, theta, Z)
Psi100 
```

```python
Z = 1
```

```python
r_vals = np.linspace(0, 20, 400)  # радиус в а.е.
Psi100_func = lambdify((r, theta, phi, Z), Psi100, modules=['numpy', {'sqrt': np.sqrt, 'exp': np.exp}])
```

```python
# 5. Строим графики
plt.plot(r_vals, Psi100_func(r_vals, 0, 0), label='1s')
plt.xlabel('r (а.о.)')
plt.grid(True)
plt.show()
```
